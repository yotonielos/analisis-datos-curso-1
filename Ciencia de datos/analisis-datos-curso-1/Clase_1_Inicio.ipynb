# Generar tablas de frecuencia absoluta y relativa

```python
#Tabla de frecuencias univariadas, comando para el uso de frecuencias absolutas
frec_data=data["nombre_categoria_producto"].value_counts().reset_index()
frec_data=frec_data.rename(columns={"count":"frecuencia_absoluta"}) #comando para cambiar el nombre de las columnas
frec_data.head(5)
```

```python
#frecuencia relativa
frec_data["porcentaje_nombre_categoria_producto"]=frec_data["frecuencia_absoluta"].apply(lambda x: x/frec_data["frecuencia_absoluta"].sum())
frec_data.head(10)
```

```python
#Calculo para la frecuencia absoluta acumulada y relativa acumulada
#calculo de la frecuencia absoluta acumulada
frec_data["acumulada_nombre_categoria_producto"]=frec_data["frecuencia_absoluta"].cumsum()
frec_data["acumulada_porcentaje_nombre_categoria_producto"]=frec_data["porcentaje_nombre_categoria_producto"].cumsum()
frec_data.head(5)
```

# Medidas de dispersión

```python
#Encontrar la desviación estandar de la longitud de los productos 
#Muestras que tan lejo de la media se encuentran las observaciones 
data["longitud_cm_producto"].std()
```

```python
#Rango intercualtil 
#Es la diferencia entre el tercer y el primer cuartil 
data["longitud_cm_producto"].quantile(0.75)-data["longitud_cm_producto"].quantile(0.25)
```

```python
#Rango, representa la diferencia que existe la observación mas grande y la más pequeña 
data["longitud_cm_producto"].max()-data["longitud_cm_producto"].min()
```

```python
#coeficiente de variación 
#Es la división de la desviación estandar sobre la media 
data["longitud_cm_producto"].std()/data["longitud_cm_producto"].mean()
```

# Correlacion de Pearson 
La correlación de Pearson es una medida estadística que cuantifica la relación lineal entre dos variables continuas, indicando su dirección y fuerza. Su valor oscila entre -1 y 1, donde 1 significa una correlación positiva perfecta, -1 una negativa perfecta y 0 ausencia de relación lineal. Es útil para identificar patrones, pero no implica causalidad. Por ejemplo, puede medir cómo se relacionan horas de estudio y calificaciones. Su fórmula se basa en la covarianza estandarizada entre las variables.

```python
#Identificar la dispersión de los datos de peso y precio
plt.scatter(data["precio"],data["peso_g_producto"])
```

```python
#Identificar la dispersión de los datos de peso y el costo del flete
plt.scatter(data["peso_g_producto"],data["valor_flete"])
```

```python
#Calcular la relación de Pearson para los gráficos vistos 
data[["peso_g_producto","valor_flete","precio"]].corr(method="pearson")
```

# Coeficiente de correlación de Spearman 
La correlación de Spearman es una medida no paramétrica que evalúa la relación monótona (no necesariamente lineal) entre dos variables, basándose en sus rangos o posiciones relativas. A diferencia de Pearson, Spearman no requiere que los datos sigan una distribución normal y es robusta ante valores atípicos. Su valor también varía entre -1 y 1:

1. 1 indica una correlación monótona positiva perfecta (si una variable aumenta, la otra también).
2. -1 señala una correlación negativa perfecta (si una aumenta, la otra disminuye).
3. 0 sugiere ausencia de relación monótona.

Es ideal para datos ordinales o cuando la relación no es lineal. Por ejemplo, analizar si el nivel educativo (ordenado por rangos) se asocia con la satisfacción laboral (también en rangos).

_*Hipótesis*_ La medida del volumen están relacionadas con el peso 

```python
#Area del producto
data["area_cm2_producto"]=data["longitud_cm_producto"]*data["altura_cm_producto"]
#Volumen del producto
data["volumen_cm3_producto"]=data["longitud_cm_producto"]*data["altura_cm_producto"]*data["ancho_cm_producto"]
data.head()
```

```python
plt.scatter(data["area_cm2_producto"],data["peso_g_producto"])
```

```python
plt.scatter(data["volumen_cm3_producto"],data["peso_g_producto"])
```

```python
#Graficar las demas relaciones alto, ancho, alto
fig, (ax1,ax2,ax3)=plt.subplots(1,3, figsize=(30,10))
fig.suptitle("Relaciones alto, ancho, alto")
ax1.scatter(data["longitud_cm_producto"],data["peso_g_producto"])
ax2.scatter(data["ancho_cm_producto"],data["peso_g_producto"])
ax3.scatter(data["altura_cm_producto"],data["peso_g_producto"])
```

```python
#Matriz de correlacion de spearman 
corr=data[["longitud_cm_producto","altura_cm_producto","ancho_cm_producto","peso_g_producto","area_cm2_producto","volumen_cm3_producto","precio",
           "valor_flete"]].corr(method="spearman")
corr
```

```python
#Graficar las demas relaciones alto, ancho, alto
fig, (ax1,ax2,ax3)=plt.subplots(1,3, figsize=(30,10))
fig.suptitle("Relaciones alto, ancho, alto")
ax1.scatter(data["longitud_cm_producto"],data["area_cm2_producto"])
ax2.scatter(data["ancho_cm_producto"],data["area_cm2_producto"])
ax3.scatter(data["altura_cm_producto"],data["area_cm2_producto"])
```

```python
#Graficar las correlaciones
corr.style.background_gradient(cmap="coolwarm").format("{:.2f}")#Indica cuantos decimales se agregan S
```
